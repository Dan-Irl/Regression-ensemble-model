```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
```

```python
# Import data
data = pd.read_csv('house_properties_data.csv') #budget is in millions for each colum
data.head()
```

```python
# set target variable to last column
data.drop(['adress'], axis=1, inplace=True)

target = data.pop('sold_price')
data['sold_price'] = target

data.head()
```

```python
# normalizing the input data using min max scaler
scaler = MinMaxScaler()
scaler.fit(data[data.columns[:-1]])

data[data.columns[:-1]] = scaler.transform(data[data.columns[:-1]])

data.info()
```

```python
data.head()
```

```python
data.to_csv('house_properties_data_normalized.csv', index=False)
```
